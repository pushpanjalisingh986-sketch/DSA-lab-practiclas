```python
#Program to Insert and Delete an Element at a Given Position in a List

list = [10,20,30,40,50]
print("original list:", list)

#Insert

pos = int(input(" enter position to insert (0-based index):"))
element = int(input("enter element to insert:"))

if pos>= 0 and pos<= len(list):
    list.insert(pos,element)
    print(" list aftre insertion:", list)
else:
    print("invalid posyion for insertion")





# Delete 
pos = int(input("enter position to delete(0-based index):"))

if pos>=0 and pos<len(list):
    deleted_element = list.pop(pos)
    print("deleted element:", deleted_element)
    print("list after deletion:",list)

else:
    print("invalid position for deletion")


```

```python
#Program to Search an Element in a List

list = [5,15,25,35,45,55]
print("list:",list)

key = int(input("enter element to search:"))

if key in list:
    print("element found at index:",list.index(key))

else:
    print("element not found")
```

```python
#Program to Demonstrate Basic Operations on Data Structures (List)

lst = []


# Insertion
lst.append(9)
lst.append(18)
lst.append(27)
print("After insertion:", lst)



# Traversal
print("Traversing list:")
for item in lst:
 print(item, end=" ")



# Searching
key = 18
if key in lst:
 print("\nElement", key, "found")
else:
 print("\nElement not found")


# Deletion
lst.remove(18)
print("After deletion:", lst)
# Length
print("Length of list:", len(lst))
```

```python

#Program to Represent Two-Dimensional Array Using Python List
# Two-dimensional array
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element at position ({i},{j}): "))
        row.append(element)
    matrix.append(row)   

print("2D Array (Matrix):")

for i in range(rows):
    for j in range(cols):
        print(matrix[i][j], end=" ")
    print()   
```
